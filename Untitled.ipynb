### Exploratory Analysis

#### GDP

Let's talk about GDP. GDP or Gross Domestic Bruto is the total of all value added both from goods and services, created in an economy. The best way to explore GDP is through a line chart because GDP data is a time series and is used to measure economic growth. 
In this chart we will explore the growth of GDP in nominal and percentage. 

```python
#Create line chart for GDP
sns.lineplot(data=df, x='year', y='gdp', hue='country')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=1)
plt.figure(figsize=(9,6))
plt.show()
```

```python
#Calculate the yoy growtth in variable 'gr'
df_pivot = df.pivot(index='year', columns='country', values='gdp')
df_pivot = df_pivot.pct_change().fillna(0).reset_index()
df_melted = pd.melt(df_pivot, id_vars=['year'], value_vars=['Chile', 'China', 'Germany', 'Mexico', 'United States of America', 'Zimbabwe'])
df_melted = df_melted.rename(columns={'value' : 'gr'})
df_melted.head()
```

```python
#Merge df_melted['gr'] into df and create line plot for economic growth ('gr')
df = df.merge(df_melted, on=('year', 'country'))
df = df.drop(columns=['gr_x', 'gr_y'])
df.head()
```

```python
sns.lineplot(data=df, x='year', y='gr', hue='country')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=1)
plt.figure(figsize=(9,6))
plt.show()
```

The best way to answer this question through Exploratory Analysis is using scatter plot graph. A scatter plot uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables.
Just like what we do to GDP and Life Expectancy, we need to split this graph based on coutries. Let's go.

But we have to take a note, scatterplots show possible associations or relationships between two variables. However, just because our graph or chart shows something is going on, it doesn’t mean that a cause-and-effect relationship exists. That's the reason we need to indentifying it using statistical tests. 

### Statistical Test 

Statistic test can be useful in data analysis and modeling to better understand the relationships between two or more variables of a dataset. Each data point in the dataset is an observation, and the features are the properties or attributes of those observations. 

#### Linear Regression

Linear regression is the process of finding the linear function that is as close as possible to the actual relationship between features. In other words, we determine the linear function that best describes the association between the features. This linear function is also called the regression line.

```python
#import libraries and conduct the test
import scipy.stats

x = df['gdp']
y = df['lifeexp']


result = scipy.stats.linregress(x, y)
result
```

Interpretation
* pvalue=0.000618712956946 < 𝛼 = 0.05. This means that there is a strong evidence of a linear association between the variables 'gdp' and 'lifeexp'
* rvalue=0.34320674844915594.  0.343 means that 34% of the variation in the variable 'lifeexp' is explained by the variable 'gdp'

#### Pearson’s Correlation 

The Pearson (product-moment) correlation coefficient is a measure of the linear relationship between two features. It’s the ratio of the covariance of the two variables divided by the product of the standard deviation of each data sample. It is the normalization of the covariance between the two variables to give an interpretable score

```python
#import libraries and conduct the test
from scipy.stats import pearsonr
corrp = pearsonr(x, y)
print(corrp)
```

We can see that the two variables are positively correlated and that the correlation is 0.343. This suggests a moderate level of correlation, e.g. a value below 0.5 and above to 0.0. 
The p-value is under 𝛼 = 0.05 that indicates the correlation is statistically significant.

#### Spearman’s  Correlation

Can we approve the Pearson's correlation?

Not yet. Two variables may be related by a nonlinear relationship, such that the relationship is stronger or weaker across the distribution of the variables. For the case of Pearson's test above, we still haven't done the normality test. 
Further, if the two variables being considered not normal. it may have a non-Gaussian distribution.

The Spearman rank-order correlation coefficient is a nonparametric measure of the monotonicity of the relationship between two datasets. Unlike the Pearson correlation, the Spearman correlation does not assume that both datasets are normally distributed.

```python
#import libraries and conduct the test
#normality test using shapiro test
from scipy.stats import shapiro
stat, p = shapiro(x)
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```

```python
stat, p = shapiro(y)
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```

Both the gdp and life expectancy are not normal distribution. 
* p <= alpha: reject H0, not normal.
* p > alpha: fail to reject H0, normal.

```python
#import libraries and conduct the test
from scipy.stats import spearmanr
corrs = spearmanr(x, y)
print(corrs)
```

From the output we can see that the Spearman rank correlation is 0.41818 and the corresponding p-value is 4.9179.
This indicates that there is a positive correlation between the gdp and life expectancy.
However, since the p-value of the correlation is not less than 0.05, the correlation is not statistically significant.
